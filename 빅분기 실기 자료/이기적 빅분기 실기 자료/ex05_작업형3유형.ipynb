```python
# 기초 통계학 검정을 다루는 내용 출제된다.
# 귀무 가설 H0는 0보다 크거나 같다.
# 대립 가설 H1은 0보다 작다.

# 출력을 원할 경우 print() 함수 활용
# 예시) print(df.head())

# getcwd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이브 경로(C: 등) 접근 불가

import pandas as pd
a = pd.read_csv('data/blood_pressure.csv', index_col=0)
# a.info()
# print(a.head())
# print(a.describe())

# 사용자 코딩
import numpy as np

# (a) 표본평균을 구하시오(반올림하여 소수 둘째자리까지 계산)
n = len(a)
answer_a = sum(a.bp_after - a.bp_before) / n # 쌍체표본은 n으로 나눠주고, 일반표본은 n-1로 나눠준다.
print(round(answer_a, 2))  # 소수점 둘째자리까지 출력
## np.mean(a.bp_after-a.bp_before) 이렇게 해도 쌍체표본은 구할 수 있다.

# (b) t값을 구하라 (반올림하여 소수 둘째자리까지 계산)
# t값
from scipy import stats
# 쌍체표본 t검정하는 코드(단, 양측검정이 아니라 단측검정이기 때문에 alternative='less'입력)
t_value, p_value = stats.ttest_rel(a.bp_after, a.bp_before, alternative='less')
# 검정 통계랑을 staticstic값, 
print(round(t_value,2))

## t값 직접 구하기 : (표본평균)/((표준편차)/root(n))
## answer_a/(np.std(a.bp_after - a.bp_before, ddof=1)/np.sqrt(n)) # ddof=1을 반드시 입력해줘야 한다.

# (c)
# p값이 0.05보다 작으니까 귀무가설 기각한다는 사실도 알 수 있음
answer_c = round(p_value, 4)
print(f"p값은 {answer_c}이므로 귀무가설을 ", end="")
if answer_c >= 0.05:
	print('채택한다.')
else : 
	print('기각한다.')
# 해당 화면에서는 제출하지 않으며, 문제 풀이 후 답안제출에서 결괏값 제출
```
