```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
# 1.크롬브라우저를 실행
driver=wb.Chrome()
# 1.1 한솥 메뉴사이트로 이동
driver.get("https://www.hsd.co.kr/menu/menu_list")
```

```python
# 상품명, 상품가격정보를 수집(글자만 출력)
# By.CSS_SELECTOR = 선택자를 직접 명시할 때 사용
# 셀렉터를 쓴 경우에는 (. #) 같은 클래스 아이디 식별자를 써눠야ㅐ 한다.
price = driver.find_elements(By.CSS_SELECTOR, ".item-price>strong")
```

```python
title = driver.find_elements(By.CLASS_NAME, "h.fz_03")
```

```python
titleList = []
for i in title:
    tData = i.text
    titleList.append(tData)
```

```python
priceList = []
for i in price:
    pData = i.text
    priceList.append(pData)
```

```python
dic = {"상품명":titleList, "상품가격":priceList}
```

```python
# 더보기 버튼을 끝까지 누르는 로직
# 선택자 한번에 뽑아오는 방법 [f12] - [네모화사표아이콘으로 원하는 거 클릭 - 파란블록] - [좌클릭 copy - copt selector] 
# 카피셀렉트는 무조건 한개 가져온다 (복수개 가져올 때는 수정을 약간 해야함)

btn = driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a")
btn.click()
```

```python
# 더보기를 반복해서 끝까지 실행!
# 반복의 횟수가 다 다르기 때문에, while문을 활용!
while True: 
    btn = driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a")
    btn.click()
    time.sleep(1)
```

# 예외처리
- 오류가 나기전까지 실행하다가 오류가 났을때는 다른코드를 실행!
- try = 오류 나기 전까지 코드
- except = 오류가 났을 때 코드

```python
try :
    # 오류 나기 전
    while True:
        btn = driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a")
        btn.click()
        time.sleep(1)
except :
    # 오류 난 후
    print("더보기 완료!")
```

# 한셀에 완벽한 크롤링 코드 짜기

```python
#1. 크롬창 실행 후 > 한솥메뉴사이트 요청
#2. 더보기 끝까지 미는 로직
#3. 가격정보, 메뉴명 수집
#4. 순수한 텍스트 정보만 추출 = 비어있는 리스트에 텍스트만 담는 작업
#5. 그 데이터를 활용하여 데이터 프레임 제작 (판다스 사용해서)
#6. 한글이 깨지지 않고 csv파일 제작

driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")

try:
    while True:
        btn=driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a")
        btn.click()
        time.sleep(1)
except:
    pass

titleList = []
priceList = []

title = driver.find_elements(By.CLASS_NAME, "h.fz_03")
price = driver.find_elements(By.CLASS_NAME, "item-price")

for i in title :
    tData = i.text
    titleList.append(tData)
for i in price:
    pData = i.text
    priceList.append(pData)
    
dic = {"상품명":titleList , "상품가격":priceList}

data1=pd.DataFrame(dic)
data1.to_csv("한솥메뉴.csv", encoding="euc-kr")
driver.close()
```

# ActionChains
- Acthion Chains 사용법
 - 1. 내가 액션을 줄 모든 요소들을 수집
 - 2. ActionChains("브라우저")호출
 - 3. ActionChains("브라우저").기능(요소명).기능(요소명).기능(요소명).perform()

```python
# 이벤트를 묶어서 실행하는 라이브러리
from selenium.webdriver import ActionChains
```

```python
driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/")
```

```python
#1. 메뉴탭에 마우스를 올리기!
#2. 올린 상태에서 전체 메뉴를 클릭!
menu = driver.find_element(By.CSS_SELECTOR, "#gnb > div.gnb_menu > ul > li:nth-child(3) > p > a")
allList = driver.find_element(By.CSS_SELECTOR, "#gnb > div.gnb_menu > ul > li:nth-child(3) > div > ul > li:nth-child(1) > a")

ActionChains(driver).move_to_element(menu).click(allList).perform()
#액션체인(내가보고 있는 드라이버). 내가 할 이벤트들 열거. perform ()동작시켜달란말
```
