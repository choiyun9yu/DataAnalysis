### BMI 데이터가 분류될 수 있는 데이터 인지 확인

```python
# matplotlib : 파이썬 그래프 시각화 도구
# Seaborn : matplotilib에 고급 그래프 시각화 도구를 제공하는 라이브러리
import matplotlib.pyplot as plt

# scatter : 산포도를 그리는 함수 (2차원 좌표에 점을 찍는 함수)
# x좌표 : Weight, y좌표 : Height
def myScatter(label, color):
    #파라미터로 넘어온 라벨의 해당 값들을 저장 - 불리안 인덱싱
    tmp = bmi[bmi["Label"] == label]
    
    #입력한 라벨 값만 산포도 그림
    #scatter(x좌표, y좌표, 색상, 라벨)
    plt.scatter(tmp["Weight"], tmp["Height"],
               c= color, label = label)

#그래프 크기 설정 (맨위에 적어야 하는 듯)
plt.figure(figsize=(10, 10))
    
#그래프 그리기
myScatter("Extreme Obesity", "purple")
myScatter("Obesity", "violet")
myScatter("Overweight", "red")
myScatter("Normal", "blue")
myScatter("Weak", "orange")
myScatter("Extremely Weak", "yellow")


# 범례 추가 (loc : 범례의 위치)
# loc를 써주지 않으면 알아서 데이터가 가장 작게 분포된 우치에 표시
plt.legend(loc="lower right")

#x축,y축 라벨표시 (기본적으로 한글은 지원하지 않음)
plt.xlabel("Weigiht")
plt.ylabel("Height")

#x축, y축 눈금의 범위
plt.xlim(0,200)
plt.ylim(0,250)

#제목 표시
plt.title("BMI data")

#격자 표시
plt.grid()

plt.show()

```

```python
#훈련데이터와 테스트데이터로 분리(훈련 70%, 테스트 30%)
X_train = bmi.iloc[0:350,1:3]
y_train = bmi.iloc[0:350,3]
X_test = bmi.iloc[350:,1:3]
y_test = bmi.iloc[350:,3]

print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

## 4. 모델링

```python
from sklearn.neighbors import KNeighborsClassifier

# n_neighbors : 이웃의 수
# 이웃의 수가 적으면 과대적합
# 이웃의 수가 많으면 과소적합
# Hyper Parameter Tuning : 적당한 이웃의 수를 결정하는 것
knn_model = KNeighborsClassifier(n_neighbors=3)
# 훈련은 훈련데이터로
knn_model.fit(X_train, y_train)  #fit은 항상 훈련데이터로만 분석
```

```python
# 정확도 계산
# 훈련정확도와 테스트 정확도가 모두 낮으면 -> 과소적합
# 훈련정확도보다 테스트 정확도가 높으면 -> 과소적합
# 훈련정확도와 테스트 정확도가 크게 차이나면 -> 과대적합
# 훈련정화도가 테스트 정확도보다 약간 높으면 -> 일반화 모델
print("훈련 정확도 : ", knn_model.score(X_train, y_train))
print("테스트 정확도 : ", knn_model.score(X_test, y_test))
```

- 이웃의 수가 감소하면 과대적합이 생김

```python
knn_model2 = KNeighborsClassifier(n_neighbors=1) # 이웃 수는 짝수를 잘 안쓴다. 어디에 가장 가까운지 정할 수 없어서
knn_model2.fit(X_train, y_train) 

print("훈련 정확도 : ", knn_model2.score(X_train, y_train))
print("테스트 정확도 : ", knn_model2.score(X_test, y_test))
```

- 이웃의 수가 증가하면 과소적합이 생김

```python
knn_model3 = KNeighborsClassifier(n_neighbors=10)
knn_model3.fit(X_train, y_train)

print("훈련 정확도 : ", knn_model3.score(X_train, y_train))
print("테스트 정확도 : ", knn_model3.score(X_test, y_test))
```

- 일반화된 모델을 찾아보자 - 하이퍼파라미터 튜닝

```python
# 이웃의 수마다의 정확도
train_acc = []
test_acc = []

#사용할 이웃의 범위 값들을 정의
neighbor = range(1,20)

for n in neighbor :
    knn_model = KNeighborsClassifier(n_neighbors=n)
    knn_model.fit(X_train, y_train)
    
    train_acc.append(knn_model.score(X_train, y_train))  #모델에 한번 학습시키고 나면 기존에 공부했던 자료
    test_acc.append(knn_model.score(X_test, y_test))
```

```python
x = train_acc
y = test_acc

plt.plot(x, label="train")
plt.plot(y, label="test")
plt.xlabel("neighbor")
plt.legend()
```

- 딕셔너리를 데이터 프레임으로 변환
- 특성과 라벨 간의 상관관계 분석

```python
import pandas as pd
df_feature = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df_feature.head()
```

```python
df_feature.info()
```

```python
df_feature.describe()
```

```python
df_label = pd.DataFrame(boston["target"], columns=["price"])

df_label
```

```python
#특성 데이터와 라벨 데이터를 하나의 데이터 프레임으로 병합 -> concat
#concat은 행에 붙으니까 열에 붙이려면 axis=1 설정 넣어줘야함
df_boston = pd.concat([df_feature, df_label], axis=1)
```

```python
# 특성과 라벨의 상관관계 분석
# .corr 상관관계 분석함수 (짝표)
df_boston.corr()
# 양의 정수 : 라벨 값과 비례관계  
# 음의 정수 : 라벨 값과 반비례관계   
# 0에 가까울 수록 집 값과 관련이 없는 것  ->  0의 상관관계는 전처리 후 사용

#다중공선성(공선: 1차함수 그래프 위에서 기울기와 y절편이 같음)
# -특성 간의 공선성(데이터 유사성) 문제  ->  특성과 특성간의 상관관계가 90% 이상인 것  
#   ->  특성을 여러개 넣는 의미가 없음  ->  특성과 라벨 간의 상관관계가 더 낮은 특성을 탈락시키면 된다
# -특성과 라벨간의 공선성(데이터 유사성) 문제  ->  특성과 라벨간의 상관관계가 90% 이상인 것 
#   ->  분석의 의미가 없음  ->  유사성이 높은 특성을 삭제하거나 처리해야 (다른 특성들과의 관계가 보임?)
#        (연역 논증의 동어반복의 오류와 비슷?)
# 0.9이상의 영향을 주는 것이 있는지 불리언 인덱싱으로 추출해보면 알아차리기 좋다.
```

- train_test_split : 훈련데이터와 테스트데이터를 분리해주는 함수
 - 디폴트: 훈련 75%, 테스트 25%로 분리


```python
from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(특성 데이터, 라벨 데이터, random_state = 7)
#random_state : 랜덤 시드를 설정 숫자 적으면 그 숫자 시드에 있는 값을 가져오겠다는 뜻
#랜덤 시드를 설정해야 항상 같은 데이터셋을 사용할 수 있음
#랜덤 데이터 셋을 어떻게 설정하느냐에 따라 성능이 달라지기도 한다.

X_train, X_test, y_train, y_test = train_test_split(boston["data"],boston["target"], random_state = 7)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 4. 모델 선택 및 훈련

```python
from sklearn.neighbors import KNeighborsRegressor

knn_model4 = KNeighborsRegressor(n_neighbors=5)
knn_model4.fit(X_train, y_train)
```

```python
#평가 (회귀분석에서는 평가도구로 오차를 사용)
#오차가 작으면 좋은 모델, 오차가 크면 안좋은 모델
#  1번 모델의 오차 : 10, -100 -> -90
#  2번 모델의 오차 : 2, 3 -> 5
# ->음수오차로 인해서 직접 비교하기 어려운 문제

# ->음수를 양수로 변경하는 방법 : 
#      제곱 -> 평균제곱오차(Mean Squared rror : MSE)   
#      절대값 -> 평균절대값오차(Mean Absolute Error : MAE)
# MSE를 더 사용하는 이유는 제곱을 하기 때문에 좋은 모델과 나쁜 모델을 구분하기 쉬워짐
# 단, MAE는 속도가 더 빨라서 가끔 사용하기도 한다.

# 통계학에서 절대값 안쓰는 이유는 미분하기 어려워져서 그렇다.

from sklearn.metrics import mean_squared_error

pred_train = knn_model4.predict(X_train)
pred_test = knn_model4.predict(X_test)

print("훈련 오차 : ", mean_squared_error(pred_train, y_train))
print("테스트 오차 : ", mean_squared_error(pred_test, y_test))
```

```python
# MSE는 제곱을 한 값이라서 원래 집값이 아님 -> 제곱근 사용
import numpy as np
print("훈련 오차 : ", np.sqrt(mean_squared_error(pred_train, y_train)))
print("테스트 오차 : ", np.sqrt(mean_squared_error(pred_test, y_test)))
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
print("훈련 오차 : ", mean_absolute_error(pred_train, y_train))
print("테스트 오차 : ", mean_absolute_error(pred_test, y_test))

#오차가 적다고 무조건 좋은게 아님(?)
#같은 모델을 가지고 다른 방식으로 오차율을 계산한 결과이기 떄문에 이 숫자가 오차율을 뜻하는 것은 아님 by 경준형님
```
