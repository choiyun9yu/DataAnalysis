## Question 45
데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

```python
# groupby 심화
## 판다스 groupby()의 기능은 데이터를 그룹별로 분할하여 독립된 그룹에 대하여 별도로 데이터를 처리(혹은 적용)하거나  
## 그룹별 통계량을 확인하고자 할 때 유용한 함수이다. 
### 1) Split 단계 : .groupby()에서 정의한 칼럼 조건에 따라 독립된 그룹으로 나누어준다. 
### 2) Apply 단계 : Split단계에서 나누어진 독립된 그룹별 원하는 함수를 적용하는 단계이다.
### 3) Combine 단걔 : 각각의 독립된 그룹별로 함수가 적용된 결과를 종합하여 다시 하나의 테이블로 합치는 단계이다.

df0 = df.groupby('host_name').size()
df0.head(5)

### 여기서는 'host_name'을 기준으로 split, size()함수를 적용, 다시 합쳐서 내놓은 결과를 상위 5개 보여준다.
```

## Question 55
데이터중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고  
neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라 

```python
# 값의 비율을 구하는 문제
sub1 = df.groupby(['neighbourhood_group', 'room_type']).room_type.size().unstack()
sub1.loc[:,:] = (sub1.values / sub1.sum(axis=1).values.reshape(-1,1))
# 두 번째 줄 코드는 계싼된 그룹 크기를 각 행의 총합으로 나누어 비율로 변환하는 작업이다. 
# sub1[:,:]를 통해 계싼된 비율로 데이터프레임의 값을 대체한다.
# axis=1은 각 행을 따라 연산을 수행하르는 의미이다.
# values.reshape(-1,1)은 행렬의 형태를 변경하여 열이 1개인 형태로 변환하는 것이다.

# reshape(-1,1)은 넘파이 배열의 형태를 변경하는 메소드이다.  
# -1은 해당 차원의 크기를 자동으로 계산하라는 의미이다.
# 괄호 앞에 행의 차원 수, 괄호 뒤에 열의 차원 수가 들어가는 여기서는 행은 자동으로 열은 1차원 이다.

sub1
```

Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라  
Unknown : N  
Less than $40K : a  
$40K - $60K : b  
$60K - $80K : c  
$80K - $120K : d  
$120 +' : e


```python
# 칼럼에서 개개인의 값에 어떻게 접근할까? apply 또는 map을 사용할거 같긴함

# map 함수 사용해서 하는 방법
dic = {
    'Unknown'           : 'N',
    'Less than $40K'    : 'a',
    '$40K - $60K'       : 'b',  
    '$60K - $80K'       : 'c',
    '$80K - $120K'      : 'd', 
    '$120K +'            : 'e',
}
df['newIncome'] = df['Income_Category'].map(lambda x : dic[x])
df['newIncome']

# map과 lambda함수 그리고 딕셔너리를 이용해서 매핑할 수 있다.

# map은 딕셔너리로 키값에 해당하는 문자열을 밸류값으로 매핑할 수 있다.
```

#### Question 58
Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라   
Unknown : N  
Less than $40K : a  
$40K - $60K : b  
$60K - $80K : c  
$80K - $120K : d  
$120K +’ : e  

```python
# apply함수 이용해서 하는 방법
# apply 메소드는 사용자 정의 함수를 각 밸류마다 적용시키는 것이다.

def changeValue(x):
    if x=='Unkown':
        return 'N'
    elif x=='Less than $40K':
        return 'a'
    elif x=='$40K - $60K':
        return 'b'
    elif x=='$60K - $80K':
        return 'c'
    elif x=='$80K - $120K':
        return 'd'
    elif x=='$120K +':
        return 'e'
    
df['newIncome'] = df.Income_Category.apply(changeValue)

df['newIncome']
```

#### Question 59
Customer_Age의 값을 이용하여 나이 구간을 AgeState 칼럼으로 정의하라.  
(0~9 : 0 , 10~19 :10 , 20~29 :20 …) 각 구간의 빈도수를 출력하라


```python
min(df['Customer_Age'])
```

```python
# 이건 apply 이용하는게 더 쉽겠다.

def to_ageStat(x):
    if x >= 0 and x < 10:
        return 0
    elif x < 20:
        return 10
    elif x < 30:
        return 20
    elif x < 40:
        return 30
    elif x < 50:
        return 40
    elif x < 60:
        return 50
    elif x < 70:
        return 60
    elif x < 80:
        return 70

df['AgeState'] = df['Customer_Age'].apply(to_ageStat)

df['AgeState'].value_counts().sort_index()
# .sort_index까지 해줘야 인덱스인 나이대 순으로 정렬 되어서 출력된다.
# .sort_index사용하지 않으면 컬럼인 빈도수가 높은 순으로 정렬 된다.
```
