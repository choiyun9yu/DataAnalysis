```r
library(dplyr)

# 깃허브에 있는 csv 파일을 읽어와서 데이터프레임 df로 넣는다.
df <- read.csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```r
df
```

```r
# Survived, Pclass, Sex, Cabin, Embarked 변수의 데이터형을 factor형으로 변환한다.
library(dplyr)
df <- mutate(df,Survived = factor(Survived))
df <- mutate(df,Pclass = factor(Pclass))
df <- mutate(df,Sex = factor(Sex))
df <- mutate(df,Cabin = factor(Cabin))
df <- mutate(df,Embarked = factor(Embarked))

df
```

```r
head(df, 10)
```

```r
str(df)
```

```r
summary(df)
```

```r
# group_by(), summarize()를 이용해서 Pclass 빈도수 구하기
# table(df@Pclass)와 동일한 결과
df %>% group_by(Pclass) %>% summarize(N = n())
```

```r
# Histogram
hist(df$Fare, freq = TRUE)
# Boxplot – Survived 값에 따른 Fare 값의 분포를 보여준다.
boxplot(formula=Fare~Survived, data=df)
```

```r
# 빈도 수 계산
print(table(df$Sex))

# 남성 생존 비율
df_m <- filter(df, df$Sex=="male")
print(prop.table(table(df_m$Survived)))

# 여성 생존 비율
df_f<- filter(df, df$Sex=="female")
print(prop.table(table(df_f$Survived)))
```
