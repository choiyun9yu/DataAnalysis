```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# 1. 브라우저를 실행 > 유튜브채널 사이트로!
driver = wb.Chrome()
driver.get("https://www.youtube.com/watch?v=N53uaeo8cdA")
```

```python
# 2. 댓글정보 수집
rpl=driver.find_elements(By.ID, "content-text")
```

```python
for i in range(len(rpl)):
    r=rpl[i].text
    print(r)
```

```python
#수집한 댓글의 길이 체크
#로딩받은 갯수에 따라서 수집된 데이터가 다르다.
#더 많은 데이터를 수집하고 싶으면 스크롤을 밑으로 내려야한다.
len(rpl)
```

# 스크롤 내리는 방법
- 키보드의 END키를 활용!
- 전체 페이지를 움직이기 때문에 body태그를 접근
- body태그에게 키보드의 END값을 전달 -> body.send_keys(Keys.END)
- [주의점] 반복 작업시 while문은 사용금지! body는 계속 존재하기 때문에 무한루프에 빠져버린다.
- [해결방법] for문을 활용해서 많은 양의 반복을 통해 진행 추천! (가장 쉬운 해결책)

```python
#end키 누르는 조건을 찾기 힘들어서 while문 쓰기 어려움
for i in range(100):
    body = driver.find_element(By.TAG_NAME, "body")
    body.send_keys(Keys.END)
    time.sleep(0.5)
```
